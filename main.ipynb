Assignment 1: Temperature Check

```python
temperature = int(input("Enter temerature: "))

if temperature < 10:
    print("It's too cold!")

elif temperature >= 10 and temperature <= 25:
    print("It's cool outside")
    
else:
    print("It's hot outside")
```

Assignment 2: ATM Withdrawal System

```python
account_balance = int(input("Enter account balance: "))
withdraw_amaount = int(input("Enter withdrawal amount: "))

if withdraw_amaount > account_balance:
    print("Insufficient Balance")

elif withdraw_amaount <= account_balance:
    print("Transaction Successful")
    print(f"Remaining Balance: {account_balance - withdraw_amaount}")
```

Assignment 3: Student Grade Calculator

```python
marks = int(input("Enter obtained marks: "))

if marks >= 90 and marks <= 100:
    print("A+")

elif marks >= 80 and marks <= 89:
    print("A")

elif marks >= 70 and marks <= 79:
    print("B") 

elif marks >= 60 and marks <= 69:
    print("C") 

else:
    print("Fail")
```

Assignment 4: Bus Ticket Discount

```python
age = int(input("Enter your age: "))

ticket_price = 1000

if age < 5:
    print("Free")

elif age >= 5 and age <= 10:

    ticket_price = ticket_price - (ticket_price * 0.5)
    print("50% discount")
    print(ticket_price)


elif age >= 60:
    print("30% discount")

else:
    print("Full price")
```

Assignment 5: Restaurant Bill with Discount

```python
bill = int(input("Enter total bill: "))

if bill > 1000:
    new_bill = bill - bill * 0.1
    print("You got 10% discount")
    print(new_bill)

else:
    print("No discount")
    print(bill)
```

Assignment 7: Traffic Light System

```python
color = input("Enter a color: ")

if color == "red":
    print("Stop")

elif color == "yellow":
    print("Ready to go")

elif color == "green":
    print("Go")

else:
    print("Invalid color")
```

Assignment 8: Leap Year Checker

```python
year = int(input("Enter year: "))

if year % 400 == 0:
    print("Leap Year")

elif year % 4 == 0 and year % 100 != 0:
    print("Leap Year")

else:
    print("Not a Leap Year")
```

Assignment 10: Electricity Bill Calculator

```python
num_units = int(input("Enter number of units: "))

bill = 0

if num_units <= 100:
    bill = num_units * 5

elif num_units >= 101 and num_units <= 200:
    bill = num_units * 7

elif num_units > 200:
    bill = num_units * 10

print(bill)
```
